The topologies are not equivalent because:
- 2 - 5 direction not preserved
- 2, 5 - 1, 3, 4 neighborhoods cannot be the same because in the case of 2 and 5 there is a spiral i.e. distance gets bigger but also slows down periodically (with 360 degrees of rotations)
- 1 - 3, 4 direction not preserved
- 3 - 4 in the case of 3 the distance between the points decreases as the time increases, while in the case of 4, it is the other way around. Moreover in the case of 4 every point passes through the steady state from the middle.
